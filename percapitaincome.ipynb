```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv("canada_per_capita_income.csv(3.2.24).csv")
print(df.to_string())
```

```python
X = df[['year']]
Y = df[['percapitaincome']]
print(X)
print(Y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
ans = model.predict([[2004]])
print(ans)
```

```python
model.score(X,Y)
```
